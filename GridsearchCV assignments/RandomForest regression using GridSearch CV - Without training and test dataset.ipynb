```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
input=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
input
```

```python
output=dataset[['charges']]
```

```python
output
```

```python
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import GridSearchCV

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

input=sc.fit_transform(input)

param_grid={'criterion':['squared_error', 'absolute_error', 'friedman_mse','poisson'],'max_features':['sqrt','log2',None],'n_estimators':[10,100]}

grid=GridSearchCV(RandomForestRegressor(),param_grid,n_jobs=-1,verbose=3,cv=5,refit=True)

grid.fit(input,output)

```

```python


y_pred=grid.predict(input)

from sklearn.metrics import r2_score

r_score=r2_score(output,y_pred)

print("The Rscore value for the best parameter{}:" .format(grid.best_params_),r_score)
```

```python
re=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(re)
```
